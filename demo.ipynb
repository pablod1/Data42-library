## Demo de prueba con algunas de las funciones de la libreria Data42

#### Para esta prueba utilizaremos el dataset de world happines del a√±o 2015

```python
import Data42.Cleaning as cl
import Data42.Learning as le
import Data42.Visualization as vi
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/happiness_2015.csv')
df['special_chars'] = '@#_45'
df.head()
```

## Cleaning

#### En este modulo podemos encontrar diversas funciones como: Limpieza de missings, la deteccion de outliers y modificar caracters en una determinada columna del dataframe

```python
cl.missings(df)
```

```python
cl.detect_outlier(df,'Freedom')
```

```python
cl.modify_characters(df,'special_chars',remove='@',add='adasdqw')
```

## Visualization


#### El modulo de visualizaci√≥n esta compuesto principalmente por graficos y reportes de informacion sobre nuestro dataframe 

```python
vi.data_report_numeric(df)
```

```python
vi.pie_random(df['Region'])
```

```python
vi.countplot('Region','Region',df)
```

## Machine Learning

#### En este modulo podemos encontrar algoritmos de prediccion y comparativas de correlacion entre nuestras variables

```python
le.balance_observation(df,'Region')
```

```python
le.correlation_coeff(df,threshold=0.5)
```
